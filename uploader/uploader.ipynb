## Setup

```python
import os
import pathlib
from supabase import create_client, Client
from dotenv import load_dotenv

env = load_dotenv()

URL = os.getenv('SUPABASE_URL')
KEY = os.getenv('SUPABASE_SECRET_KEY')
BUCKET = "stills"
IMPORT_ROOT = pathlib.Path.cwd().joinpath("works")

DRY_RUN = False
KEEP_ORIGINAL = False

supabase: Client = create_client(supabase_url=URL or "", supabase_key=KEY or "")
```

## Images

```python
from scripts.files_helper import import_work_folder


work_dirs = sorted([p for p in IMPORT_ROOT.iterdir() if p.is_dir()])
for work_dir in work_dirs:
    import_work_folder(
        work_dir,
        supabase,
        BUCKET,
        dry_run=DRY_RUN,
        keep_original=KEEP_ORIGINAL,
    )
```
